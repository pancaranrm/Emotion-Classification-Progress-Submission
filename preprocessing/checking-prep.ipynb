```python
import pandas as pd

train_df = pd.read_csv('../dataset/train_ready_final.csv')
val_df = pd.read_csv('../dataset/val_ready.csv')
test_df = pd.read_csv('../dataset/test_ready.csv')

# 1. Ringkasan Jumlah Data
print("--- Ringkasan Jumlah Data ---")
print(f"Total Data Training (Balanced) : {len(train_df)} baris")
print(f"Total Data Validation (Original): {len(val_df)} baris")
print(f"Total Data Testing (Original)   : {len(test_df)} baris")
print("-" * 30)

# 2. Distribusi Emosi pada Data Training (Sudah Seimbang)
print("\n--- Distribusi Emosi (Data Training) ---")
print(train_df['emotion'].value_counts())

# 3. Distribusi Emosi pada Data Testing (Representasi Dunia Nyata)
print("\n--- Distribusi Emosi (Data Testing) ---")
print(test_df['emotion'].value_counts())

# 4. Melihat Contoh Perubahan Teks (Preprocessing Check)
print("\n--- Sampel Teks Sebelum vs Sesudah Preprocessing ---")
# Menampilkan 3 sampel dari data training
sample = train_df[['caption', 'caption_cleaned', 'emotion']].head(3)
for i, row in sample.iterrows():
    print(f"\n[Emosi: {row['emotion']}]")
    print(f"Original : {row['caption'][:100]}...")
    print(f"Cleaned  : {row['caption_cleaned'][:100]}...")
```
