```python
import pandas as pd

# 1. Membaca kedua file CSV
df1 = pd.read_csv('dataset/Twitter_Data_Final_20260113_231315.csv')
df2 = pd.read_csv('dataset/Twitter_BIGDATA_5K_20260114_000844.csv')

# 2. Menggabungkan kedua dataframe secara vertikal
# ignore_index=True digunakan agar index direset menjadi urutan baru
df_merged = pd.concat([df1, df2], ignore_index=True)

# 3. Opsional: Menghapus duplikat berdasarkan kolom unik seperti 'uuid'
df_merged = df_merged.drop_duplicates(subset=['uuid'])

# 4. Menyimpan hasil penggabungan ke file CSV baru
df_merged.to_csv('merged_twitter_data.csv', index=False)

print(f"Berhasil menggabungkan data!")
print(f"Jumlah baris file 1: {len(df1)}")
print(f"Jumlah baris file 2: {len(df2)}")
print(f"Total baris setelah digabung: {len(df_merged)}")
```
